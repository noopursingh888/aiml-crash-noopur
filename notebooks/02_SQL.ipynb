```python
query = """
SELECT COUNT(*) AS Total_Orders
FROM orders;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT payment_type,
COUNT(*) AS Total_Orders
FROM orders
GROUP BY payment_type;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT AVG(payment_value) AS Avg_Payment
FROM orders;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT customer_state,
COUNT(*) AS Total_Orders
FROM orders
GROUP BY customer_state
ORDER BY Total_Orders DESC;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT COUNT(*) AS Delayed_Orders
FROM orders
WHERE delayed = 1;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT customer_state,
AVG(estimated_delay_days) AS Avg_Delay
FROM orders
GROUP BY customer_state
ORDER BY Avg_Delay DESC;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT order_id,
payment_value
FROM orders
ORDER BY payment_value DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT order_month,
COUNT(*) AS Total_Orders
FROM orders
GROUP BY order_month;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT AVG(freight_value) AS Avg_Freight
FROM orders;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT *
FROM orders
WHERE payment_value >
(
SELECT AVG(payment_value)
FROM orders
);
"""

pd.read_sql(query, conn)
```
